```python
from pathlib import Path
from pydub import AudioSegment
import os
```

```python
def split_audio(input, output_dir, chunk_size_ms):
    audio_path = input
    ## dividir el audio en una cantidad de milisegundos especifico
    audio = AudioSegment.from_file(audio_path)
    chunks = [audio[i:i + chunk_size_ms] for i in range(0, len(audio), chunk_size_ms)]
    for i, chunk in enumerate(chunks):
        formatted_string = f"{i:04d}"
        chunk.export(os.path.join(output_dir, f"chunk_{formatted_string}.wav"), format="wav")
    return chunks
```

```python
def list_files_in_directory(directory):
    path = Path(directory)
    return [str(p) for p in path.rglob("*") if p.suffix in (".wav")]
```

```python
def check_for_chunk_in_filenames(directory):
    path = Path(directory)
    for file in path.rglob("*"):
        if "chunk" in file.name:
            return True
    return False
```

```python
output_dir = "../output/13 de Setiembre/1.Radio/fase_1/"
chunk_size_ms = 50000
all_files = list_files_in_directory(output_dir)
all_files = [file for file in all_files if (".wav" in file) & ("chunk" not in file)]
```
