```python
from pydub import AudioSegment
from google.cloud import speech_v1p1beta1 as speech
import os
from pathlib import Path
import time
```

```python
def list_files_in_directory(directory, pattern):
    path = Path(directory)
    return [str(p) for p in path.rglob("*") if p.suffix in (pattern)]
```

```python
def transcribe_mono_audio(audio_path, text_path, credentials_path):
    ## speech to text gcp api
    ## para wav es 44100
    
    client = speech.SpeechClient.from_service_account_file(credentials_path)    
    config = speech.RecognitionConfig(
        encoding=speech.RecognitionConfig.AudioEncoding.LINEAR16,
        sample_rate_hertz=48000,
        language_code="es-ES",
    )
    with open(audio_path, "rb") as audio_file:
        audio_content = audio_file.read()

    audio = speech.RecognitionAudio(content=audio_content)
    response = client.recognize(config=config, audio=audio)

    ## escribir el resultado
    index = 0
    for result in response.results:
        resultado = result.alternatives[0].transcript
        ##with open('../output/audio.txt', 'a') as file:
        with open(text_path, 'a') as file:
            file.write(resultado + "\n")
        print(resultado)
```

```python
credentials_path = "../input/proyecto-nlp-credenciales.json" #path de las credenciales gcp
output_dir = "../output/13 de Setiembre/1.Radio/fase_1/"
all_files = list_files_in_directory(output_dir, ".wav")
all_files = [file for file in all_files if (".wav" in file) & ("chunk" in file)]

all_dir = []
for index, f in enumerate(all_files):
    dir_name = "\\".join(all_files[index].split("\\")[:-1])
    all_dir.append(dir_name)
all_dir = list(set(all_dir))
```

```python
all_files_text = list_files_in_directory(output_dir, ".txt")
all_files_text = [file for file in all_files_text if (".txt" in file)]

all_dir_text = []
for index, f in enumerate(all_files_text):
    dir_name_text = "\\".join(all_files_text[index].split("\\")[:-1])
    all_dir_text.append(dir_name_text)
all_dir_text = list(set(all_dir_text))
```

```python
dir_disponibles = list(set(all_dir) - set(all_dir_text))
```

```python
start_time = time.time()
for dir in dir_analizar:
    print(dir)
    text_path = dir + f'/audio_a_texto.txt'
    all_specific_file_in_dir = [file for file in all_files if (dir in file)]
    for audio in all_specific_file_in_dir:
        transcribe_mono_audio(audio, text_path, credentials_path)
end_time = time.time()
elapsed_time = end_time - start_time
print(f"Code execution time: {elapsed_time} seconds")
```
