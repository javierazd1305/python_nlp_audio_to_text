```python
from pydub import AudioSegment
from google.cloud import speech_v1p1beta1 as speech
import os
from pathlib import Path
import openai
import time
import json
```

```python
def list_files_in_directory(directory, pattern):
    path = Path(directory)
    return [str(p) for p in path.rglob("*") if p.suffix in (pattern)]
```

```python
def transcribe_mono_audio(audio_path, text_path):
    audio_file= open(audio_path, "rb")
    transcript = openai.Audio.transcribe("whisper-1", audio_file)
    text = transcript["text"]
    ## escribir el resultado
    with open(text_path, 'a') as file:
        file.write(text + "\n")
    return text
```

```python
all_files = list_files_in_directory(output_dir, ".mp3")
all_files = [file for file in all_files if (".mp3" in file) & ("chunk" in file)]
all_dir = []
for index, f in enumerate(all_files):
    dir_name = "\\".join(all_files[index].split("\\")[:-1])
    all_dir.append(dir_name)
all_dir = list(set(all_dir))
```

```python
all_files_text = list_files_in_directory(output_dir, ".txt")
all_files_text = [file for file in all_files_text if (".txt" in file)]
all_dir_text = []
for index, f in enumerate(all_files_text):
    dir_name_text = "\\".join(all_files_text[index].split("\\")[:-1])
    all_dir_text.append(dir_name_text)
all_dir_text = list(set(all_dir_text))
```
